```python
import tabula
import tabulate
```

```python
from tabula import read_pdf
from tabulate import tabulate

#reads table from pdf file
df = read_pdf("text-and-table.pdf",pages="all",multiple_tables=True) #address of pdf file
if df:
    for i, table in enumerate(df): # loop that iterates overs the full pdf and find out tables
        print(f"Table {i+1}:")
        print(tabulate(table,headers='keys',tablefmt='grid')) # provides the output in well-structured format.
        print("\n")
else:
    print("No tables")
```

```python
# converted extracted tables to json
import json
if df:
    json_rep={}
    for i,table in enumerate(df):
        table_json = table.to_json(orient='records') # each row is json obj(converts dataframe to json string)
        json_rep[f'Table_{i+1}'] = json.loads(table_json) #converst json string to python dict for ease in reading.
    print(json.dumps(json_rep,indent=4))
else:
    print('no table')
```
