```python
# collecting data through scraping data from a website from wikipedia. 

#importing the libraries
import requests
from bs4 import BeautifulSoup
import spacy
import pandas as pd

#  Scrape data from a website
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser") # html content of the website

# Extract text from paragraphs
text = "\n".join([p.text for p in soup.find_all("p")])

# Load SpaCy model and tokenize the text
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

# Take a sample of tokens (first 10 tokens for demonstration purposes)
bool_attributes = []
for attr in dir(doc[0]):
    try:
        value = getattr(doc[0], attr)
        # Check if it's a boolean attribute or a method that returns a boolean
        if isinstance(value, bool) or (callable(value) and isinstance(value(), bool)):
            bool_attributes.append(attr)
    except Exception:
        # Skip attributes/methods that cause errors
        continue

# Create a dictionary with boolean values for each method across all tokens
data = {attr: [] for attr in bool_attributes}
for token in doc:
    for attr in bool_attributes:
        value = getattr(token, attr)() if callable(getattr(token, attr)) else getattr(token, attr)
        data[attr].append(value)

# Convert to DataFrame
df = pd.DataFrame(data, index=[token.text for token in doc]).T

# Rename columns for better readability
df.columns = [token.text for token in doc]


df
```
