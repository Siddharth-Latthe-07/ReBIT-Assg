```python
# Gender Dist
plt.figure(figsize = (6,4))
sns.countplot(data = user_data,x='gender',palette='Set2')
plt.title('gender distribution')
plt.show()
```

```python
# Age Dist
user_data['age'] = user_data['current_age']
plt.figure(figsize = (8,6))
sns.histplot(user_data['age'], kde=True, bins =30,color = 'blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.histplot(user_data['credit_score'],kde=True,bins=30,color='green')
plt.title('credit score dist')
plt.xlabel('credit score')
plt.ylabel('Freq')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=cards_data,x='has_chip',palette='Set2')
plt.title('cards with chip vs no chip')
plt.show()
```

```python
merged_data = pd.merge(user_data,cards_data,left_on='id',right_on='client_id',how='inner')
merged_data.drop(columns=['id_x'],axis=1,inplace=True)
merged_data.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged_data,x='yearly_income',y='credit_limit',hue='gender',palette='viridis')
plt.title("income vs credit limit")
plt.xlabel('yearly income')
plt.ylabel('credit limit')
plt.show()
```
