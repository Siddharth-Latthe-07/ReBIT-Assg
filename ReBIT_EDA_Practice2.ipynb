```python
# Gender Dist
plt.figure(figsize = (6,4))
sns.countplot(data = user_data,x='gender',palette='Set2')
plt.title('gender distribution')
plt.show()
```

```python
# Age Dist
user_data['age'] = user_data['current_age']
plt.figure(figsize = (8,6))
sns.histplot(user_data['age'], kde=True, bins =30,color = 'blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=cards_data,x='has_chip',palette='Set2')
plt.title('cards with chip vs no chip')
plt.show()
```

```python
merged_data = pd.merge(user_data,cards_data,left_on='id',right_on='client_id',how='inner')
merged_data.drop(columns=['id_x'],axis=1,inplace=True)
merged_data.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged_data,x='yearly_income',y='credit_limit',hue='gender',palette='viridis')
plt.title("income vs credit limit")
plt.xlabel('yearly income')
plt.ylabel('credit limit')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(user_data['per_capita_income'],kde=True,bins=30,color='blue')
plt.title('dist of per capita income')
plt.xlabel('per capita income')
plt.ylabel('Freq')
plt.show()
```

```python
user_data['age_group'] = pd.cut(user_data['current_age'],bins=[0,25,40,60,80],labels=['Young(<25)','Mid-age(25-40)','senior(40-60','Elderly(>60)'])
plt.figure(figsize=(12,6))
sns.countplot(data=user_data,x='age_group',palette='pastel')
plt.title('age grp dist')
plt.show()
```

```python
user_data['year_group'] = pd.cut(user_data['birth_year'],bins=[1920,1940,1960,1980,2000],labels=['1920-1940','1940-1960','1960-1980','1980-2000'])
plt.figure(figsize=(12,6))
sns.countplot(data=user_data,x='year_group',palette='pastel')
plt.title('year grp dist')
plt.show()
```
